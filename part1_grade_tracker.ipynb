```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:

    name = student["name"].strip().title()
    roll = int(student["roll"])

    marks = []
    for m in student["marks_str"].split(", "):
        marks.append(int(m))

    valid = True
    for word in name.split():
        if not word.isalpha():
            valid = False

    if valid:
        status = "Valid name"
    else:
        status = "Invalid name"

    card = (name + " → " + status + "\n" +
            "================================\n" +
            "Student : " + name + "\n" +
            "Roll No : " + str(roll) + "\n" +
            "Marks   : " + str(marks) + "\n" +
            "================================\n")

    print(card)

    cleaned_students.append({"name": name, "roll": roll, "marks": marks})

for student in cleaned_students:
    if student["roll"] == 103:
        output = "Special Output for Roll 103:\n" + student["name"].upper() + "\n" + student["name"].lower()
        print(output)
```

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

print("Student:", student_name)

for i in range(len(subjects)):
    m = marks[i]
    if m >= 90:
        grade = "A+"
    elif m >= 80:
        grade = "A"
    elif m >= 70:
        grade = "B"
    elif m >= 60:
        grade = "C"
    else:
        grade = "F"
    print(subjects[i], ":", m, "→", grade)


    total = 0
for m in marks:
    total = total + m

average = round(total / len(marks), 2)

highest_mark = marks[0]
highest_sub  = subjects[0]
lowest_mark  = marks[0]
lowest_sub   = subjects[0]

for i in range(len(marks)):
    if marks[i] > highest_mark:
        highest_mark = marks[i]
        highest_sub  = subjects[i]
    if marks[i] < lowest_mark:
        lowest_mark = marks[i]
        lowest_sub  = subjects[i]

print("Total   :", total)
print("Average :", average)
print("Highest :", highest_sub, "-", highest_mark)
print("Lowest  :", lowest_sub,  "-", lowest_mark)

new_count = 0

while True:
    subject_input = input("Enter subject name (or 'done' to stop): ")

    if subject_input.lower() == "done":
        break

    marks_input = input("Enter marks for " + subject_input + ": ")

    if not marks_input.isnumeric():
        print("Warning: Invalid input. Please enter a number.")
    else:
        m = int(marks_input)
        if m < 0 or m > 100:
            print("Warning: Marks must be between 0 and 100.")
        else:
            subjects.append(subject_input)
            marks.append(m)
            new_count = new_count + 1
            print(subject_input, "added successfully!")

new_total = 0
for m in marks:
    new_total = new_total + m

new_average = round(new_total / len(marks), 2)

print("New subjects added :", new_count)
print("Updated Average    :", new_average)

```
